```python
import matplotlib.pyplot as plt
import numpy as np
pi = 3.141592654
```

```python
sample_data = [_*_ for _ in range(0,1000) ]
print(len(sample_data))
```

```python
x = np.linspace(0,2*pi,10000)
# print(x)
z = np.sin(x)
# print(z)
```

```python
fig, ax = plt.subplots()
ax.plot(x, z)


ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()
```
