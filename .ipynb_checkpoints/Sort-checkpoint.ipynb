```python
def BubbleSort(input_list):
    #copy the list. do not sort in place
    input_l = list(input_list)
    
    for _ in range(len(input_l)-1):
        for i in range(len(input_l)-1): 
            #compares the left element to the one next to it
            if input_l[i]>input_l[i+1]:
                #swap the two
                input_l[i], input_l[i+1] = input_l[i+1], input_l[i]
                pass
                
    return input_l
```

```python
def InsertSort(input_list):
    input_l = list(input_list)
    
    for i in range(len(input_l)):
        current = input_l.pop(i) #this is the current value that will be moved
        j=i-1
        
        while j >= 0 and current < input_l[j]:
            j-=1
            
        input_l.insert(j+1,current)
        #print(input_l)
        
    return input_l
```

```python
def MergeSort(input_list):
    input_l = list(input_list)
    
    if len(input_l)<=1:
        return input_l
    
    mid = int(len(input_l)/2)
    left = input_l[:mid]
    right = input_l[mid:]

    left, right = MergeSort(left), MergeSort(right)
    
    merged_list = Merge(left,right)
    
    return merged_list
```

```python
def Merge(l, r):
    new_list = [] #this
   
    l_iter, r_iter = 0,0

    while l_iter < len(l) and r_iter < len(r):
        if l[l_iter] <= r[r_iter]:
            new_list.append(l[l_iter])
            l_iter +=1
        else:
            new_list.append(r[r_iter])
            r_iter +=1
            
    return new_list + l[l_iter:] + r[r_iter:]
```

```python
def QuickSort(input_list):
    
    #returns the first position in the array that has the minimum value of the array
    def argmin(array):
        list_l = list(array)
        mini = min(list_l)

        for iter, i in enumerate(array):
            if i == mini:
                return iter
        return -1
    
    #copy the list i guess
    input_l = list(input_list)
    #first need to pick a value to compare to the rest of. let's choose the last one
    
    if len(input_l) <= 1:
        return input_l
    
    val = input_l[-1]    
    left = [y for y in input_l[:-1] if y <val]
    right = [x for x in input_l[:-1] if x>=val]

    if len(left) == 0 and len(right) == 0:
        #this loop does literally nothing lmao
        left = []
        right = []
        
    elif len(left) == 0 and len(right) >1:
        #need to take some of the right
        #mid = int(len(right)/2)
        
        #still taking things out of order, because in the case of 
        # [] 0 [5,1,1,1]
        # it becomes
        # [0] 5 [1,1,1] which obviously shoudn't happen
        left = [val]
        val = min(right) #take the minimum value
        argm = argmin(right)
#         right = right[:argm] + right[argm+1:]
        right.pop(argm)


    elif len(right) == 0 and len(left) >1:
        #mid = int(len(left)/2)
        
        #need to take some the left
        right = [val]
        val = min(left)
        argm = argmin(left)
#         left = left[:argm] + left[argm+1:]
        left.pop(argm)
        
        
    print(left, val, right)
    left_sort, right_sort = QuickSort(left), QuickSort(right)
    #print(left)
    #print(right)

    
    return left_sort + [val] + right_sort
```

```python
def QuickSort2(input_list):
    #duplicate the list
    input_l = list(input_list)
    
    if len(input_l) == 1:
        return input_l
    
    pivot_num = input_l[0]#choose the end of it
    left = [x for x in input_l if x<=pivot_num]
    right = [x for x in input_l if x>pivot_num]
    
    print(left, pivot_num, right)
    
    left, right = QuickSort2(left), QuickSort2(right)
    
    return left + [pivot_num] + right
    
```
